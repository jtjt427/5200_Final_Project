```python
import pandas as pd
```

```python
df1 = pd.read_csv("IEA-EV-dataEV salesHistoricalCars.csv")
df2 = pd.read_csv("IEA-EV-dataEV stockHistoricalCars.csv")
```

```python
# Creating new columns for each powertrain type and assigning values based on conditions
df1['BEV'] = df1['value'].where(df1['powertrain'] == 'BEV')
df1['PHEV'] = df1['value'].where(df1['powertrain'] == 'PHEV')
df1['EV'] = df1['value'].where(df1['powertrain'] == 'EV')

# Calculating the average values for each powertrain type
bev_average = df1['BEV'].mean()
phev_average = df1['PHEV'].mean()
ev_average = df1['EV'].mean()

bev_average, phev_average, ev_average
```

```python
# Grouping by 'Year' and 'Entity', summing the values for BEV, PHEV, and EV
grouped_data = df1.groupby(['Year', 'Entity']).agg({
    'BEV': 'mean',
    'PHEV': 'mean',
    'EV': 'mean'
}).reset_index()

# Displaying the updated grouped data
grouped_data
```

```python
# Grouping the data by 'Year' and dropping the 'powertrain' column
grouped_data = df1.drop('powertrain', axis=1).groupby('Year').mean()
grouped_data.reset_index(inplace=True)
grouped_data.head(6)
```

```python
df3 = pd.read_csv("IEA-EV-dataEV salesHistoricalCars.csv")
```

```python
# Splitting the 'parameter' column into separate columns for each parameter type
pivot_data = df3.pivot_table(index=['Year'], columns='parameter', values='value', aggfunc='mean')

# Resetting index to make 'Year' a column again
pivot_data.reset_index(inplace=True)
pivot_data.head(6)
```

```python
full_data = pd.merge(grouped_data, pivot_data, on=['Year'])
full_data
```

```python
new_df = full_data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
new_df
```

```python
merged_data = pd.merge(data, full_data, on=['Entity', 'Year'], how='left')

# Display the first few rows of the merged dataframe to ensure the merge was successful
merged_data.head()
# merged_data.to_csv("final_sales.csv")
```

```python
import pandas as pd
import plotly.graph_objs as go

df = merged_data
df['Year'] = df['Year'].astype(str)

def get_info(x):
    return x['Year'] + " : " + "{:.2f}".format(x["Electric cars sold"])

df["info"] = df.apply(get_info, axis=1)

grouped_df = df.groupby(['Entity', 'Code', 'Continent']).agg({
    'Electric cars sold': 'sum', 
    'info': lambda x: '<br>'.join(x)
}).reset_index()

traces = []
buttons = []
continents = grouped_df['Continent'].unique()

for continent in continents:
    continent_data = grouped_df[grouped_df['Continent'] == continent]
    trace = go.Choropleth(
        locations=continent_data['Code'],
        z=continent_data['Electric cars sold'],
        colorscale='Viridis',
        colorbar=dict(title='Number'),
        zmin=0,
        zmax=max(grouped_df['Electric cars sold']),
        showscale=True,
        reversescale=True,
        hoverinfo='text',
        text=[f"{name}<br>{other_info}<br>Total: {sales}"
              for name, other_info, sales in zip(continent_data['Entity'], continent_data['info'], continent_data['Electric cars sold'])],
        visible = (continent == continents[0]) # Only the first continent is visible initially
    )
    traces.append(trace)
    
    # Button for the current continent
    buttons.append(dict(
        label=continent,
        method='update',
        args=[{'visible': [continent == c for c in continents]},
              {'title': f'Number of new electric cars sold 2010-2022 in {continent}',
               'showlegend': False}]
    ))

layout = go.Layout(
    title='Number of new electric cars sold 2010-2022',
    geo=dict(
        showcoastlines=False,
        showframe=False,
        projection=dict(type='natural earth')
    ),
    updatemenus=[{
        'buttons': buttons,
        'direction': 'down',
        'showactive': True,
    }]
)

fig = go.Figure(data=traces, layout=layout)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Load and prepare the data
# df = pd.read_csv("IEA-EV-dataEV stockHistoricalCars.csv")
df5 = pd.read_csv("filtered.csv")

# Filter for unique values for dropdowns
parameters = df5['parameter'].unique()
powertrains = df5['powertrain'].unique()

# Initial plot setup
initial_parameter = parameters[0]
initial_powertrain = powertrains[0]

# Filter data for initial plot
filtered_df = df5[(df5['parameter'] == initial_parameter) & (df5['powertrain'] == initial_powertrain)]

# Create the figure
fig = make_subplots()

for region in filtered_df['region'].unique():
    Entity_df = filtered_df[filtered_df['region'] == region]
    fig.add_trace(go.Scatter(x=Entity_df['year'], y=Entity_df['value'], mode='lines', name=region))

# Update layout for initial plot
fig.update_layout(title=f'{initial_parameter} - {initial_powertrain}', xaxis_title='Year', yaxis_title='Value')

# Create dropdowns for parameters and powertrains
param_dropdown = {
    'buttons': [
        {
            'label': param,
            'method': 'update',
            'args': [{'visible': [p == param for p in df5['parameter']]}]
        } for param in parameters
    ],
    'direction': 'down',
    'showactive': True,
}

powertrain_dropdown = {
    'buttons': [
        {
            'label': powertrain,
            'method': 'update',
            'args': [{'visible': [p == powertrain for p in df5['powertrain']]}]
        } for powertrain in powertrains
    ],
    'direction': 'down',
    'showactive': True,
}

# Add dropdowns to the figure
fig.update_layout(
    updatemenus=[
        {
            'buttons': param_dropdown['buttons'],
            'direction': param_dropdown['direction'],
            'showactive': param_dropdown['showactive'],
            'x': 0.0,
            'xanchor': 'left',
            'y': 1.1,
            'yanchor': 'top'
        },
        {
            'buttons': powertrain_dropdown['buttons'],
            'direction': powertrain_dropdown['direction'],
            'showactive': powertrain_dropdown['showactive'],
            'x': 0.15,
            'xanchor': 'left',
            'y': 1.1,
            'yanchor': 'top'
        }
    ]
)

# Show the figure
fig.show()
```

This plot shows a good view for each parameter and powertrain. people are able to click on the dropdown list and visulize their favorite field. Here, different categories refer to differet results, for example, for electricity demand, China has a huge EV demand, but on the other hand, Germany has a huge BEV demand, and the Europe and the US seems has a higher PHEV demand. No matter PHEV, BEV, or EV. The demand are increase over time.
