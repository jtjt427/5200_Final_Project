```python
df1['BEV'] = df1['value'].where(df1['powertrain'] == 'BEV')
df1['PHEV'] = df1['value'].where(df1['powertrain'] == 'PHEV')
df1['EV'] = df1['value'].where(df1['powertrain'] == 'EV')

bev_average = df1['BEV'].mean()
phev_average = df1['PHEV'].mean()
ev_average = df1['EV'].mean()

bev_average, phev_average, ev_average
```

```python
grouped_data = df1.groupby(['Year', 'Entity']).agg({
    'BEV': 'mean',
    'PHEV': 'mean',
    'EV': 'mean'
}).reset_index()

grouped_data
```

```python
import pandas as pd
import plotly.graph_objs as go

df = merged_data
df['Year'] = df['Year'].astype(str)

def get_info(x):
    return x['Year'] + " : " + "{:.2f}".format(x["Electric cars sold"])

df["info"] = df.apply(get_info, axis=1)

grouped_df = df.groupby(['Entity', 'Code', 'Continent']).agg({
    'Electric cars sold': 'sum', 
    'info': lambda x: '<br>'.join(x)
}).reset_index()

traces = []
buttons = []
continents = grouped_df['Continent'].unique()

for continent in continents:
    continent_data = grouped_df[grouped_df['Continent'] == continent]
    trace = go.Choropleth(
        locations=continent_data['Code'],
        z=continent_data['Electric cars sold'],
        colorscale='Viridis',
        colorbar=dict(title='Number'),
        zmin=0,
        zmax=max(grouped_df['Electric cars sold']),
        showscale=True,
        reversescale=True,
        hoverinfo='text',
        text=[f"{name}<br>{other_info}<br>Total: {sales}"
              for name, other_info, sales in zip(continent_data['Entity'], continent_data['info'], continent_data['Electric cars sold'])],
        visible = (continent == continents[0])
    )
    traces.append(trace)
    
    buttons.append(dict(
        label=continent,
        method='update',
        args=[{'visible': [continent == c for c in continents]},
              {'title': f'Number of new electric cars sold 2010-2022 in {continent}',
               'showlegend': False}]
    ))

layout = go.Layout(
    title='Number of new electric cars sold 2010-2022',
    geo=dict(
        showcoastlines=False,
        showframe=False,
        projection=dict(type='natural earth')
    ),
    updatemenus=[{
        'buttons': buttons,
        'direction': 'down',
        'showactive': True,
    }]
)

fig = go.Figure(data=traces, layout=layout)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# df = pd.read_csv("IEA-EV-dataEV stockHistoricalCars.csv")
df5 = pd.read_csv("filtered.csv")

parameters = df5['parameter'].unique()
powertrains = df5['powertrain'].unique()

initial_parameter = parameters[0]
initial_powertrain = powertrains[0]

filtered_df = df5[(df5['parameter'] == initial_parameter) & (df5['powertrain'] == initial_powertrain)]

fig = make_subplots()

for region in filtered_df['region'].unique():
    Entity_df = filtered_df[filtered_df['region'] == region]
    fig.add_trace(go.Scatter(x=Entity_df['year'], y=Entity_df['value'], mode='lines', name=region))

fig.update_layout(xaxis_title='Year', yaxis_title='Value')

param_dropdown = {
    'buttons': [
        {
            'label': param,
            'method': 'update',
            'args': [{'visible': [p == param for p in df5['parameter']]}]
        } for param in parameters
    ],
    'direction': 'down',
    'showactive': True,
}

powertrain_dropdown = {
    'buttons': [
        {
            'label': powertrain,
            'method': 'update',
            'args': [{'visible': [p == powertrain for p in df5['powertrain']]}]
        } for powertrain in powertrains
    ],
    'direction': 'down',
    'showactive': True,
}

fig.update_layout(
    updatemenus=[
        {
            'buttons': param_dropdown['buttons'],
            'direction': param_dropdown['direction'],
            'showactive': param_dropdown['showactive'],
            'x': 0.0,
            'xanchor': 'left',
            'y': 1.1,
            'yanchor': 'top'
        },
        {
            'buttons': powertrain_dropdown['buttons'],
            'direction': powertrain_dropdown['direction'],
            'showactive': powertrain_dropdown['showactive'],
            'x': 0.15,
            'xanchor': 'left',
            'y': 1.1,
            'yanchor': 'top'
        }
    ]
)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Load the CSV file
df_final_sales = pd.read_csv('final_sales.csv')  # Make sure to provide the correct path to the file

# Create the figure
fig_sales = make_subplots()

# Add a trace for each country for the 'BEV' metric
for country in df_final_sales['Entity'].unique():
    country_df = df_final_sales[df_final_sales['Entity'] == country]
    fig_sales.add_trace(go.Scatter(
        x=country_df['Year'], 
        y=country_df['BEV'],  # Assuming 'BEV' is the name of the column you want to plot
        mode='lines', 
        name=f"{country} - BEV",
        visible=True  # Set all 'BEV' traces to be visible
    ))

# Update layout for initial plot
fig_sales.update_layout(title='BEV by Country', xaxis_title='Year', yaxis_title='BEV')

# Metrics for dropdown (excluding 'BEV' since it's already visible)
metrics = ['EV', 'PHEV', 'EV sales', 'EV sales share', 'EV stock', 
           'EV stock share', 'Electricity demand', 'Oil displacement Mbd', 
           'Oil displacement, million lge']

# Create a button for each metric, and update the visibility of the traces
buttons = []
for i, metric in enumerate(metrics):
    visibility = [metric == m for country in df_final_sales['Entity'].unique() for m in metrics]
    button = dict(
        label=metric,
        method='update',
        args=[{'visible': visibility},
              {'title': f'{metric} by Country',
               'yaxis': {'title': metric}}],
    )
    buttons.append(button)

# Add dropdown to the figure
fig_sales.update_layout(
    updatemenus=[{
        'buttons': buttons,
        'direction': 'down',
        'showactive': True,
        'x': 0.0,
        'xanchor': 'left',
        'y': 1.1,
        'yanchor': 'top'
    }]
)

# Show the figure
fig_sales.show()
```
